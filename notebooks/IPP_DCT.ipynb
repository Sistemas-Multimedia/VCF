# IPP_DCT: Hybrid video coding using IPP motion compensation and DCT

```python
VIDEO = "mobile_352x288x30x420x300.mp4"
ENCODED_PREFIX = "encoded_"
DECODED_PREFIX = "decoded"
```

```python
%%bash -s "$VIDEO"

VIDEO=$1

if [[ -f "/tmp/$VIDEO" ]]; then
    echo "The file exists."
else
    echo "The file does not exist."
    wget "http://www.hpca.ual.es/~vruiz/videos/$VIDEO" -O "/tmp/$VIDEO" --quiet
fi
```

### Using YCrCb

```python
%%bash -s "$VIDEO" "$ENCODED_PREFIX"

VIDEO=$1
ENCODED_PREFIX=$2

python ../src/IPP_DCT.py encode -i "/tmp/$VIDEO" -O $ENCODED_PREFIX -N 20 -G 10 -M 16 -S 8 -q 64 -R 0.1 \
  --color_transform YCrCb
```

```python
%%bash -s "$ENCODED_PREFIX" "$DECODED_PREFIX"

ENCODED_PREFIX=$1
DECODED_PREFIX=$2

python ../src/IPP_DCT.py decode -i $ENCODED_PREFIX -O $DECODED_PREFIX -N 20 -G 10 -M 16 -q 64 \
  --color_transform YCrCb
```

```python
%%bash -s "$VIDEO" "$ENCODED_PREFIX"  "$DECODED_PREFIX"

VIDEO=$1
ENCODED_PREFIX=$2
DECODED_PREFIX=$3

MIN_COEF=-2048
MAX_COEF=2047
QSS=64

python ../src/IPP_DCT.py encode \
  -i "/tmp/$VIDEO" \
  -O $ENCODED_PREFIX \
  -N 20 \
  -G 10 \
  -M 16 \
  -S 8 \
  -q 64 \
  -R 0.1 \
  --quantizer LloydMax -m $MIN_COEF -n $MAX_COEF -q $QSS


echo ""
echo "=========================================="
echo "DECODING"
echo "=========================================="

python ../src/IPP_DCT.py decode \
  -i $ENCODED_PREFIX \
  -O $DECODED_PREFIX \
  -N 20 \
  -G 10 \
  -M 16 \
  -q 64 \
  --quantizer LloydMax -m $MIN_COEF -n $MAX_COEF -q $QSS
```
