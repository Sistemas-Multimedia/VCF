# IPP_DCT: Hybrid video coding using IPP motion compensation and DCT

```python
%%bash

python ../src/main.py
```

```python
%%bash

VIDEO="mobile_352x288x30x420x300.mp4"

if [[ -f "/tmp/$VIDEO" ]]; then
    echo "The file exists."
else
    echo "The file does not exist."
    wget "http://www.hpca.ual.es/~vruiz/videos/$VIDEO" -O "/tmp/$VIDEO" --quiet
fi

OUTPUT_PREFIX="encoded_"

python ../src/IPP_DCT.py encode \
  -i "/tmp/$VIDEO" \
  -O $OUTPUT_PREFIX \
  -N 20 \
  -G 10 \
  -M 16 \
  -S 8 \
  -q 64 \
  -R 0.1 \
  --st 2D-DCT \
  --color_transform YCrCb \
  --quantizer LloydMax -m -1024 -n -1023 -q 32


echo ""
echo "=========================================="
echo "DECODING"
echo "=========================================="

python ../src/IPP_DCT.py decode \
  -i $OUTPUT_PREFIX \
  -O ${OUTPUT_PREFIX}_dec \
  -N 20 \
  -G 10 \
  -M 16 \
  -q 64 \
  --st 2D-DCT \
  --color_transform YCrCb \
  --quantizer LloydMax -m -1024 -n -1023 -q 32

echo ""
echo "=========================================="
echo "COMPLETED"
echo "=========================================="
echo "Output files are in: ../src/_tmp/"
echo "Reconstructed video: ../src/_tmp/${OUTPUT_PREFIX}_dec.mp4"
```
