```python
%%bash
# Codificacin
python ../src/NLM.py encode -i /tmp/original.png -o /tmp/original.nlm

# Decodificacin aplicando el filtro NLM con fuerza 10
python ../src/NLM.py decode -i /tmp/original.nlm -o /tmp/decoded.png -f NLM -s 8
```
